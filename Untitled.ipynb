```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('GDP_capita.csv')
```

```python
country_name_list = []
country_code_list = []
year_list = []
gdp_capita_list = []

# df.columns[4:len(df)-1]

for index, row in df.iterrows():
    country_name = row['Country Name']
    country_code = row['Country Code']
    for year in df.columns[4:len(df)-1]:
        if not np.isnan(row[year]):
            country_name_list.append(country_name)
            country_code_list.append(country_code)
            year_list.append(year)
            gdp_capita_list.append(row[year])         
```

```python
new_df = pd.DataFrame({'Country Name':country_name_list, 'Country Code': country_code_list, 'Year': year_list,
                       "GDP Capita": gdp_capita_list})
```

```python
new_df.to_csv('GDP_capita_new.csv')
```
